```python
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib notebook
```

```python
data = pd.read_csv("./data/NFLX.csv", index_col=0, parse_dates=True)
```

```python
data.head()
```

```python
#note that volume is in miliions o nothing else shows up....this is a bad way to do it obviously
data.plot()
```

```python
fig, ax = plt.subplots()
data['Close'].plot(ax=ax)
ax.set_ylabel("Price")
ax.set_title("Netflix Close")
```

```python
#if you look at help, rows, cols are default of 1 (so you get one graph like above)
#set it at 2 and 2 gives you 4 graphs
fix, ax = plt.subplots(2, 2)
# ax is an array of an array so you access sub array
data['Open'].plot(ax=ax[0, 0], title="Open")
data['High'].plot(ax=ax[0, 1], title="High")
data['Low'].plot(ax=ax[1, 0], title="Low")
data['Close'].plot(ax=ax[1, 1], title="Close")

#spaces the graphs out
plt.tight_layout()
```

```python
fix, ax = plt.subplots()
# Limit time to summer of 2021
#choose column and then rows
data['Volume'].loc['2021-06-01':'2021-08-15'].plot.bar(ax=ax, title="Volume")


```
