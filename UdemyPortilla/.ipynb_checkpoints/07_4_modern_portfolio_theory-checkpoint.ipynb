Modern Portfolio Theory: Given the same returns, investors would chose the portfolio with less risk

Lets build up our portfolio of 4 stocks

First way: Equal Weight (25% each)
But how do we decide other % to max return and min risk

Could optimze based on historical data using random avgs-----monte carlo
basically plot all of those and find optimal version (bullet curve).

The x axis is vol and y is returns. So if you trace the top of the curve that is the 
Efficient Frontier----max return for that desired vol

There are optimization techniques as well that do not rely on monte carlo




```python
import numpy as np
import pandas_datareader as pdr
import datetime as dt
import matplotlib.pyplot as plt
import pandas as pd
```

```python
start = dt.datetime(2017, 1, 1)
end = dt.datetime(2022, 3, 10)
#Costco, Walmart, Target, Dollar General
COST = pdr.get_data_yahoo("COST", start, end)['Adj Close']
WMT = pdr.get_data_yahoo("WMT", start, end)['Adj Close']
TGT = pdr.get_data_yahoo("TGT", start, end)['Adj Close']
DG = pdr.get_data_yahoo("DG", start, end)['Adj Close']

```

```python
frames = [COST, WMT, TGT, DG]

retail = pd.concat(frames, axis=1)

retail.columns = ['COST', 'WMT', 'TGT', 'DG']
```

```python
retail
```

```python
#make a new df with daily pct change of returns
retail_returns = retail.pct_change(1).dropna()
```

```python
retail_returns
```

```python
retail['COST'].plot()
```

```python
retail_returns['TGT'].plot(kind='hist', bins=100)
```

```python
#we want to look at our pct return at any given time (i.e. from time we bought until x time)
#in other words we want the y axis to be our cumulative returns
#start at $1.00
```

```python
cumul_return = (1 + retail_returns['COST']).cumprod() - 1
```

```python
cumul_pct_return_costco = 100 * cumul_return
```

```python
cumul_pct_return_costco.plot()
```

```python
#dot product----multiply index 0 times index 0 of 2 df, etc. 
#then you cna get return of each stock in one array, and the weight in a second
```

```python
#assume we put 100% into costco
all_in_costco = [1, 0 , 0, 0]
```

```python
#This gives us daily returns of the portfoli (which should match costco obviously)
np.dot(all_in_costco, retail_returns.transpose())
```

```python
retail_returns['COST']
```

```python
#equal weights in all 4
N = len(retail_returns.columns)
N
```

```python
equal_weights = N * [1/N]
equal_weights
```

```python
equal_weight_returns = np.dot(equal_weights, retail_returns.transpose())
```

```python
equal_weight_returns
```

```python
cum_equal_weight_returns_pct = ((1+equal_weight_returns).cumprod()-1)*100
```

```python
cum_equal_weight_returns_pct_df = pd.Series(cum_equal_weight_returns_pct, index=retail_returns.index)
```

```python
cum_equal_weight_returns_pct_df
```

```python
cum_equal_weight_returns_pct_df.plot(title="Eq Weight Cum Returns")
```

```python
# Lets compare equal weights versus all in on costco
```

```python
plt.figure(figsize=(10,3), dpi=200)
cum_equal_weight_returns_pct_df.plot(label="Eq Weight Cum Returns", legend=True)
cumul_pct_return_costco.plot(label="Costco", legend=True)
```
