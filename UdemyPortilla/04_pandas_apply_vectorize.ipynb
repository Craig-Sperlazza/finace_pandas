```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/home/craig/Documents/python_finance/UdemyPortilla/Python-Finance-QuantConnect/03-Core-Pandas/tips.csv')
df
```

```python
#example: grab the last four digits of credit card number
def last_four(num):
    return int(str(num)[-4:])
```

```python
#notice that you pass in the function but you dont call it
#essentially pandas calls it on each row
df['last_four'] = df['CC Number'].apply(last_four)
```

```python
df
```

```python
#lets create a yelp-like function
df['total_bill'].mean()
```

```python
def yelp(price):
    if price < 12.00:
        return '$'
    elif price > 12.00 and price < 20.00:
        return '$$'
    else:
        return '$$$'
```

```python
df['Yelp'] = df['total_bill'].apply(yelp)
```

```python
df
```

```python
#Use lambda funcs to use apply method on multiple columns
lambda x: x*2
df['double_tip'] = df['tip'].apply(lambda x: x*2)
```

```python
df
```

```python
#create a generous column if the tip to total bill meets a threshold
def quality(total_bill, tip):
    if tip / total_bill > 0.20:
        return "Generous"
    else:
        return "Cheap"
```

```python
df['tip_quality'] = df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis=1)
```

```python
df
```

```python
#vectorize is faster

df['Vector Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
```
