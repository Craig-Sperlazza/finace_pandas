```python
#20 day rolling avg
df['20mva'] = df['Adj Close'].rolling(20).mean()
df
```

```python
df['1 day change'] = df['Adj Close'].diff(1)
df['%change(1 day)'] = df['Adj Close'].pct_change(1)
df
```
