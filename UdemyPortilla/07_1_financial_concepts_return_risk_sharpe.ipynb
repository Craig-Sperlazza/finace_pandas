Measurements of Return

Pure dollar gains do not reflect cost to purchase
So you have to normalize the returns by expressing the PERCENTAGE GAIN

Percent Gain = (final - initial)/initial) * 100

100 to 400 is 300 gain and 300% Percent Gain

3000 to 3300 is 300 gain and only 10% Percent Gain


To fairly compare times: use mean daily percent return
we can use mean daily percent return as a normalized metric between different holdings for different periods of time

Need to think about diustribution for longer/shorter periods

Measurements of Risk

Many types of financial risk---see wikipedia

We will focus of measuring risk of security or bond in a portfolio
risk is defined as the variance or std dev of a holding/portfolio





One way to calculate the risk is the std dev of daily percent returns
.std()

Sharpe Ratio---takes into account both risk and return to measure investments against each other (reward to variability)

risk free rate (treasury bond yields)---recent history these have basically been zero
    SO WE ARENT INCLUDING THEM----BUT BE AWARE IF THEY RISE

SR = (Expected Return - Risk Free Rate) / std dev  of Return

#in terms of actual daily returns
SR = Mean Asset Return / Std Dev. of Assset Returns

NOTE: WANT A HIGH SHARPE RATIO (HIGH RETURN / LOW STD DEV)
GENERALLY:
    SR 0-1 is suboptimal (1 is the SP500)
    SR 1-2 is good
    SR 2-3 is very good
    SR >3 is excellent


Typically use a yearly range---252 trading days
std dev is the square root of variance:
    so multiply our dialy sharpe ratio by sqrt(252) to get yearly sharpe ration

```python
import pandas_datareader as pdr
import datetime as dt
import numpy as np
```

```python
#NOTE: we are going to get an extra day so we can drop it so we dont have a NaN on first day
#when you do the daily pct change
start = dt.datetime(2021, 2, 26)
end = dt.datetime(2022, 3, 1)
```

```python

aapl = pdr.get_data_yahoo("AAPL", start, end)
nflx = pdr.get_data_yahoo("NFLX", start, end)
```

```python
def compute_sharpe_ratio(df, risk_free_rate=0):
    mean_return = df['Daily Return'].mean()
    std = df['Daily Return'].std()
    sharpe_ratio = (mean_return - risk_free_rate)/std
    
    annual_sharpe_ratio = sharpe_ratio * (np.sqrt(252))
    
    return annual_sharpe_ratio
```

```python
sharpe_nflx = compute_sharpe_ratio(nflx)
```

```python
sharpe_aapl = compute_sharpe_ratio(aapl)
```
